# Chapter 3 - It Starts with a Tensor

## Part II: Named Tensors

```python
import torch
```

```python
_ = torch.tensor([0.2126, 0.7152, 0.0722])
_
```

```python
img_t = torch.rand(3, 5, 5) # shape: [channels, rows, columns]
img_t
```

```python
weights = torch.tensor([0.2126, 0.7152, 0.0722])
weights
```

```python
batch_t = torch.rand(2, 3, 5, 5) # shape: [batch, channels, rows, columns]
batch_t
```

```python
img_gray_naive = img_t.mean(-3)
img_gray_naive
```

```python
batch_gray_naive = batch_t.mean(-3)
batch_gray_naive
```

```python
weights
```

```python
unsqueezed_weights = weights.unsqueeze(-1)
unsqueezed_weights
```

```python
unsqueezed_weights = unsqueezed_weights.unsqueeze_(-1)
unsqueezed_weights
```

```python
img_weights = (img_t * unsqueezed_weights)
img_weights
```

```python
batch_weights = (batch_t * unsqueezed_weights)
batch_weights
```

```python
img_gray_weighted = img_weights.sum(-3)
img_gray_weighted
```

```python
batch_gray_weighted = batch_weights.sum(-3)
batch_gray_weighted
```
